```python
import numpy as np 
import pandas as pd 
from pathlib import Path
```

*1. Examening the Physical Hieroglyphics Dataset:*
Note: Images are derived from high-quality photographs taken within the Pyramid of Unas.

```python
dataset = Path("../data/original/unas_dataset")

count = []

for class_folder in dataset.iterdir():
    if class_folder.is_dir():
        image_count = len(list(class_folder.glob("*")))
        count.append([class_folder.name, image_count])

count.sort(reverse=True, key=lambda x: x[1])

for class_name, image_count in count:
    print(f"{class_name}: {image_count}")
```
